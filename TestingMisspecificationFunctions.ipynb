# Testing the effect of breaking the linear drift

```python
def b1(power: float):
    return lambda x: -x**power

powers = [3, 5, 7, 11]


drifts = [b1(p) for p in powers]
diffProsses = [DiffusionProcess(d, sigma) for d in drifts]

x_range_invariant = np.linspace(-5, 5, 100)
x_range_xi = np.linspace(0, 1, 100)
invariant_densities = [pros.invariant_density(x_range_invariant) for pros in diffProsses]
expected_hitting_times = [pros.xi(x_range_xi) for pros in diffProsses]

fig, ax = plt.subplots(1,2,figsize=(16,5))

for i, invariant_density in enumerate(invariant_densities):
    ax[0].plot(x_range_invariant, invariant_density, label=f"power1 = {powers[i]}" )
ax[0].legend(loc="upper left")
ax[0].set_title("Invariant density")
for i, expected_hitting_time in enumerate(expected_hitting_times):
    ax[1].plot(x_range_xi, expected_hitting_time, label=f"power1 = {powers[i]}")
ax[1].legend(loc="upper left")
ax[1].set_title("Expected hitting times")
ax[1].ticklabel_format(style="plain")
#ax[1].set_ylim([0, 100])
plt.show()
```

# Plotting chosen ones together

```python
def b1(power: float):
    return lambda x: -x**power

def b3(power: float):
    return lambda x: -x**power + x**2+x

def b5(power: float):
    return lambda x: -x**power + x

powers = [3, 13]
driftFuncs = ["b1", "b2", "b3"]

drifts = [b1(p) for p in powers] + [b3(p) for p in powers] + [b5(p) for p in powers]
diffProsses = [DiffusionProcess(d, sigma) for d in drifts]

x_range_invariant = np.linspace(-3, 3, 200)
x_range_xi = np.linspace(0, 1.5, 200)
invariant_densities = [pros.invariant_density(x_range_invariant) for pros in diffProsses]
expected_hitting_times = [pros.xi(x_range_xi) for pros in diffProsses]
```

```python
# Define colors for each drift function
colors = {
    "b1": "#66c2a5",
    "b2": "#fc8d62",
    "b3": "#8da0cb"
}

fig, ax = plt.subplots(1,2,figsize=(8.3, 3.2))

for i, invariant_density in enumerate(invariant_densities):
    power = powers[i % 2]
    driftFunc = driftFuncs[i // 2]
    linestyle = '--' if power == 13 else '-'
    color = colors[driftFunc]
    ax[0].plot(x_range_invariant, invariant_density, label=f"Drift power = {power}, {driftFunc}", linestyle=linestyle, color=color)
    
ax[0].set_title("Invariant density", fontsize=13, fontweight="bold")
ax[0].tick_params(axis='both', which='major', labelsize=11)
ax[0].tick_params(axis='both', which='minor', labelsize=11)

for i, expected_hitting_time in enumerate(expected_hitting_times):
    power = powers[i % 2]
    driftFunc = driftFuncs[i // 2]
    linestyle = '--' if power == 13 else '-'
    color = colors[driftFunc]
    ax[1].plot(x_range_xi, expected_hitting_time, label=f"power = {power}, {driftFunc}", linestyle=linestyle, color=color)

ax[1].set_title("Expected hitting times", fontsize=13, fontweight="bold")
ax[1].ticklabel_format(style="plain")
ax[1].tick_params(axis='both', which='major', labelsize=11)
ax[1].tick_params(axis='both', which='minor', labelsize=11)
#ax[1].set_ylim([0, 100])

# Create custom legend for line styles
line_style_solid = mlines.Line2D([], [], linestyle='-', color='black', label='3')
line_style_dashed = mlines.Line2D([], [], linestyle='--', color='black', label='13')
legend_style = ax[0].legend(handles=[line_style_solid, line_style_dashed], loc='upper right', title='Power')
ax[0].add_artist(legend_style)

# Create custom legend for colors
color_legend_elements = [mlines.Line2D([], [], color=colors[driftFunc], label=f"{driftFunc}") for driftFunc in driftFuncs]
legend_color = ax[1].legend(handles=color_legend_elements, loc='upper left', title='Drift Function')
ax[1].add_artist(legend_color)
ax[1].set_ylim([0, 15])
plt.tight_layout()
#plt.savefig(r'C:\Users\razz_\OneDrive - Aarhus Universitet\UNI\DataVidenskab\10. Semester\Images\Misspecification\DensityAndHittingTimes2.pdf')
plt.show()
```

```python
def b1(power: float):
    return lambda x: -x**power

def b3(power: float):
    return lambda x: -x**power + x**2+x

def b5(power: float):
    return lambda x: -x**power + x

powers = [3, 13]
driftFuncs = ["b1", "b2", "b3"]

drifts = [b1(p) for p in powers] + [b3(p) for p in powers] + [b5(p) for p in powers]

vals = np.linspace(-1.5,1.5,200)
Xs = [[b(x) for x in vals] for b in drifts]
```

```python
# Define colors for each drift function
colors = {
    "b1": "#66c2a5",
    "b2": "#fc8d62",
    "b3": "#8da0cb"
}

fig, ax = plt.subplots(1,1,figsize=(8.3, 3.2))

for i, X in enumerate(Xs):
    power = powers[i % 2]
    driftFunc = driftFuncs[i // 2]
    linestyle = '--' if power == 13 else '-'
    color = colors[driftFunc]
    ax.plot(vals, X, label=f"power = {power}, {driftFunc}", linestyle=linestyle, color=color)
    
ax.set_title("Misspecified drift functions", fontsize=13, fontweight="bold")
ax.tick_params(axis='both', which='major', labelsize=11)
ax.tick_params(axis='both', which='minor', labelsize=11)
ax.ticklabel_format(style="plain")

# Create custom legend for line styles
line_style_solid = mlines.Line2D([], [], linestyle='-', color='black', label='3')
line_style_dashed = mlines.Line2D([], [], linestyle='--', color='black', label='13')
legend_style = ax.legend(handles=[line_style_solid, line_style_dashed], loc='upper right', title='Power')
ax.add_artist(legend_style)

# Create custom legend for colors
color_legend_elements = [mlines.Line2D([], [], color=colors[driftFunc], label=f"{driftFunc}") for driftFunc in driftFuncs]
legend_color = ax.legend(handles=color_legend_elements, loc='lower left', title='Drift Function')
ax.add_artist(legend_color)
# ax.axhline(y=0, xmin=-1.5, xmax=1.5, color="black")
# ax.vlines(x=0, ymin=-5, ymax=5, color="black")
ax.set_ylim([-2.5, 2.5])
plt.tight_layout()
#plt.savefig(r'C:\Users\razz_\OneDrive - Aarhus Universitet\UNI\DataVidenskab\10. Semester\Images\Misspecification\DriftFunctions.pdf')
plt.show()
```
